```python
import pandas as pd
import numpy as np

# Set seed untuk hasil reproducible
np.random.seed(42)

# Jumlah data
n = 1000

# Generate data
data = pd.DataFrame({
    'Usia': np.random.randint(18, 65, size=n),
    'Jenis_kelamin': np.random.choice(['Laki_Laki', 'Perempuan'], size=n),
    'Penghasilan': np.random.normal(5000000, 500000, size=n).astype(int),
    'Pendidikan_Terakhir': np.random.choice(['SMA', 'S1', 'S2', 'S3'], size=n),
    'Status': np.random.choice(['Jomblo', 'Menikah', 'Cerai'], size=n),
    'Status_Pekerjaan': np.random.choice(['Bekerja', 'Tidak_Bekerja', 'Siswa'], size=n),
    'Jumlah_pinjaman': np.abs(np.random.normal(10000000, 30000000, size=n).astype(int)),
    'Jenis_pinjaman': np.random.choice(['KPR', 'Kendaraan', 'Konsumtif', 'Modal_Usaha'], size=n),
    'Kunjungan_Website': np.random.poisson(5, size=n),
    'Hasil': np.random.choice(['Disetujui','Ditolak'], size=n)  # Target untuk klasifikasi
})

data.head()
data.to_csv('PinjamanBank.csv',index=False)
```

```python
# Informasi dasar tentang dataset
data.info()

# Statistik deskriptif untuk kolom numerik
data.describe()

# Distribusi data untuk kolom kategorikal
for kolom in ['Usia', 'Pendidikan_Terakhir', 'Status', 'Status_Pekerjaan', 'Jenis_pinjaman']:
    print(f"\nDistribusi kolom {kolom}:\n", data[kolom].value_counts())
```
