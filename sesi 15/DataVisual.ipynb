```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

bulan = ['januari','februari','maret','april','mei','juni','juli','agustus','september','oktober','november','desember']
#pengeluaran antar 500.000-2.000.000

minggu_1 = np.random.randint(500000,2000001,12)
minggu_2 = np.random.randint(500000,2000001,12)
minggu_3 = np.random.randint(500000,2000001,12)
minggu_4 = np.random.randint(500000,2000001,12)
print(minggu_1)
data = {
    'bulan' : bulan,
    'minggu 1' : minggu_1,
    'minggu 2' : minggu_2,
    'minggu 3' : minggu_3,
    'minggu 4' : minggu_4,
}

df = pd.DataFrame(data)
df['total_pengeluaran'] = df[['minggu 1','minggu 2','minggu 3','minggu 4']].sum(axis=1)
df.describe()
```

```python
#total pengeluaran per bulan

plt.figure(figsize=(12,6))#canvas dengan ukuran widhtm height, (inch)
plt.title('total pengeluaran per bulan')
plt.bar(df['bulan'],df['total_pengeluaran'],color='red')
plt.xticks(rotation=45)
plt.ylabel("rupiah")
plt.grid(axis='y')
plt.tight_layout()
plt.show
```

```python
#scatter plat untuk analisa data pengeluaran per minggu per bulan
plt.figure(figsize=(12,6))
for minggu in ['minggu 1','minggu 2','minggu 3','minggu 4']:
    plt.plot(df['bulan'],df[minggu],marker='o',label=minggu)
plt.title("pengeluaran per minggu")
plt.xlabel('bulan')
plt.ylabel('pengeluaran (rupiah)')
plt.legend()
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#pie chart -> prosentase
plt.figure(figsize=(10,10))
plt.pie(df['total_pengeluaran'],labels=df['bulan'],startangle=90,autopct='%1.1f%%')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

df['bulan_angka'] = np.arange(1,13)

X = df[['bulan_angka']]
y = df['total_pengeluaran']

model = LinearRegression()

model.fit(X,y)

prediksi = model.predict([[13]])[0]
print(f"hasil prediksi adalah : {prediksi:,.0f}")
```
